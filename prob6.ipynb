```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 1000

# Generate two independent exponential random variables
U1 = np.random.exponential(scale=1, size=num_samples)
U2 = np.random.exponential(scale=1, size=num_samples)

# Generate X_i by adding U1 and U2
X = U1 + U2

# Plot the histogram
plt.hist(X, bins=30, density=True, alpha=0.7, color='blue', label='Histogram')

# Plot the theoretical density function f_X(x) = xe^(-x)
x_values = np.linspace(0, 8, 100)
fx_values = x_values * np.exp(-x_values)
plt.plot(x_values, fx_values, color='red', label='f_X(x)')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Histogram and Theoretical Density Function')
plt.legend()

# Show the plot
plt.show()
```
