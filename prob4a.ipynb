```python
# (a) Let X be a discrete random variable whose pmf is given by PX(X = i) = i/10 , 
# i = 1, 2, 3, 4. Generate X1, X2, . . . , X1000 i.i.d. ∼ PX using
# randint command. Verify if the generated values are according to PX using histogram

import numpy as np
import matplotlib.pyplot as plt

y=[]
x=np.random.randint(1,11,1000)
for i in range(len(x)):
    if x[i] <= 1:
        y.append(1)
    elif x[i] <=3:
        y.append(2)
    elif x[i] <=6:
        y.append(3)
    else:
        y.append(4)
plt.hist(y,bins=10,density=True)

plt.title('Histogram of X')
plt.xlabel('X values')
plt.ylabel('Probability density')
plt.show()
```
